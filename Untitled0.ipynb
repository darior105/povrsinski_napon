```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#unosim podatke
G = np.array([58.5321, 60.922, 61.41, 61.65, 63.497])
T = np.array([74.9, 65.7, 56.2, 47.5, 40.6])
```

```python
#LR
result = stats.linregress(T, G)

slope = result.slope
intercept = result.intercept
std_err = result.stderr
intercept_stderr = result.intercept_stderr

# izbaci slope, intercept, standard errors
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Standard error of slope: {std_err}")
print(f"Standard error of intercept: {intercept_stderr}")
```

```python

T_line = np.linspace(min(T), max(T), 100)
G_line = slope * T_line + intercept


plt.figure(figsize=(8, 6))
plt.scatter(T, G, color='blue', label='Data points')
plt.plot(T_line, G_line, 'r-', label=f'Linear fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Temperature (T)')
plt.ylabel('Function G(T)')
plt.title('Linear Regression of G vs. T')
plt.legend()
plt.show()
```
